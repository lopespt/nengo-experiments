```python
import sklearn.manifold as sk
import sklearn.cluster as sk3
import sklearn.decomposition as sk2
import sklearn.datasets as data
import matplotlib.pyplot as plt
import matplotlib as mplot
import numpy as np
import networkx as nx
```

```python
digits = data.load_digits(n_class=10)
figs = digits.data.reshape(-1,8,8)
colunas = 10
linhas = 10

img = np.zeros([8*linhas,8*colunas])
for i in range(linhas):
    for j in range(colunas):
        img[i*8:(i+1)*8, j*8:(j+1)*8, ] = figs[i*colunas+j,:,:]
plt.figure(figsize=(10,5))
plt.imshow(img ,cmap='Blues',aspect='auto') 
#mplot.colors.cnames
plt.show()
```

```python

#m = sk.Isomap(n_neighbors=10,n_jobs=3)
m = sk2.KernelPCA(n_components=2)
r = m.fit_transform(digits.data)

plt.scatter(r[:,0], r[:,1], c=digits.target, cmap='terrain')
plt.show()

```
