```python
import mpl_tookits
import pandas as p
import matplotlib.pyplot as plt
import mpl_tookits
import sklearn.cluster as cl
import sklearn.decomposition as dec
```

```python
t = p.read_excel(r"/Users/wachs/Dropbox/FEI/Graduacao/CC5661 - Analise de Algoritmos/estatisticas/2016A.xlsx","P1")
```

```python
t=t[:37]
a=t.groupby('1')
a=a['1'].count()
a.plot()
a=t.groupby('2')
a=a['2'].count()
a.plot()
a=t.groupby('3')
a=a['3'].count()
a.plot()
plt.show()
plt.figure()
t['1'].hist()
plt.grid(False)
plt.show()
```

```python
c = cl.hierarchical.AgglomerativeClustering(4)
a = c.fit_predict(t[['1','2','3','4','5','6']].as_matrix())
```

```python
t['grupo'] = a
```

```python
t.sort_values(by=['1','2'])
pontos = t[['1','2','grupo']].as_matrix()

plt.scatter(pontos[:,0], pontos[:,1], c=t.index, s=[ (a+b)*100 for a,b in pontos[:,[1,2]] ],alpha=0.4 )
plt.show()
```

```python
p = dec.pca.PCA(3)
p=p.fit_transform(t[ ['1', '2','3', '4', '5', '6']].as_matrix())
plt.scatter(p[:,0], p[:,1],c=t[['grupo']].as_matrix())

plt.show()
```
