```python
import sklearn.datasets as sk
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sk.load_iris()
```

```python
X = data.data
Y = data.target
```

```python
X = np.random.random(1800)*100
X = X.reshape(-1,1).astype(np.float64)
print(X.shape)
Y = 9.7*np.sin(0.3*X)+10#+np.random.normal(size=(18000))
plt.scatter(X,Y)
plt.show()
```

```python
x = tf.placeholder(tf.float32, [None, 1.0])
W = tf.Variable(tf.random_uniform_initializer([1,1800], dtype=tf.float64), name="W")
b = tf.Variable(tf.zeros([1.0], dtype=tf.float64), name="B")
init = tf.global_variables_initializer()
y = tf.nn.softmax(tf.matmul(X,W) + b)
```

```python
loss = tf.reduce_sum(tf.square(y-Y))
optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
s = tf.Session()
s.run(init)
```

```python
losses = []
for step in range(10):
    s.run(train)
    losses.append(s.run(loss))

print(s.run([W, b]))
```

```python
plt.plot(losses)
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(X, s.run(y),'+r')
plt.show()
```
